```python
# def convert(obj):
#     L=[]
#     for i in obj:
#         L.append(i['name'])
#     return L   
import ast

def convert(obj):
     L=[]
     for i in ast.literal_eval(obj):
         # print(i)
         L.append(i['name']) # i ke ander se name nikal liya ..i ki value is tarha hai {'id': 28, 'name': 'Action'}
     return L  
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
     L=[]
     counter=0
     for i in ast.literal_eval(obj):
         if counter!=3:
            L.append(i['name']) # i ke ander se name nikal liya ..i ki value is tarha hai {'id': 28, 'name': 'Action'}
            counter+=1
         else:
             break
     return L 
```

```python
movies['cast']=movies['cast'].apply(convert3)
print(movies.cast)
```

```python
import ast

def fetch_director(obj):
     L=[]
     
     for i in ast.literal_eval(obj):
         if i['job']=='Director': # jab mera job director ke == hoga tabhi ye append karega
            L.append(i['name']) # i ke ander se name nikal liya ..i ki value is tarha hai {'id': 28, 'name': 'Action'}
              
     return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview'] =movies['overview'].apply(lambda x:x.split())# x:me value aati hai x.split() operation perform hota hai
```

```python
# movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])

movies = movies.copy()
movies.loc[:, 'genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies.loc[:, 'keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies.loc[:, 'cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies.loc[:, 'crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+ movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))# :" ".join(x)=hume har list ko space se join karna hai
```

```python
# simlarity find karni hai
# text ko vector me convert kana hai
# Bang of wave
# tfidf
# word of vector

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
# features ke sath khelna hai ab
# loved,loving,love =love
# love,love,love    =   love


from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
ps =PorterStemmer()
wl =WordNetLemmatizer()

# print(ps.stem("lovly"))
# print(wl.lemmatize("lovly"))
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

def lemmatize(text):
    y =[]
    for i in text.split():
        y.append(ps.lemmatize(i))
    return " ".join(y)
```

```python
new_df['tags'].apply(stem)
```

```python
# har movie ka dusri movie ke sath distance calculate karna hai
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
similarity[0]
```

```python
list(enumerate(similarity[0])) # 0 movie ki simalarity aa rahi hai other movies ke sath
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])# lambda x:x[1] ->second bale no ke base prr simalarity find karna hai 
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
